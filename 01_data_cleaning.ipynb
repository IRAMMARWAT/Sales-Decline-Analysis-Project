## Project Overview
This project analyzes a 4-year sales dataset (2018-2021) to identify root causes of declining sales and provide actionable recommendations. The analysis follows a structured approach from data cleaning to strategic recommendations.

## Business Problem
A retail company has experienced declining sales over the past 2 years. Management needs to understand:
- Which regions and categories are underperforming

- What strategic changes can reverse the decline

# # Cell 4: Data Cleaning

````python
import numpy as np
````

````python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")
````

````python
# Cell 4: Data Cleaning
"""
## 🧹 DATA CLEANING OPERATIONS
Handling missing values, duplicates, and data type conversions
"""

# 1. Check and remove duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
df = df.drop_duplicates()
print(f"Shape after removing duplicates: {df.shape}")

# 2. Handle missing values
# For numerical columns, fill with median
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)

# For categorical columns, fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown', inplace=True)

# 3. Convert date columns
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format="mixed", dayfirst=True)

# 4. Create derived columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['DayOfWeek'] = df['Order Date'].dt.dayofweek
df['Quarter'] = df['Order Date'].dt.quarter
df['WeekOfYear'] = df['Order Date'].dt.isocalendar().week

# 5. Calculate profit margin
# Assuming the 'Profit' column is missing and you have a 'Sales' column
df['Profit'] = df['Sales']  # Replace 'Sales' with the actual column name if different

# Now you can calculate the 'Profit_Margin' column
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100

# 6. Create sales category
df['Sales_Category'] = pd.cut(df['Sales'], 
                               bins=[0, 50, 200, 500, 1000, float('inf')],
                               labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

print("\n✅ Data cleaning completed!")
print(f"Final dataset shape: {df.shape}")
print(f"Date range: {df['Order Date'].min()} to {df['Order Date'].max()}")
````
