```python
# Cell 1: Setup for Time Series
"""
## ⏰ TIME SERIES FORECASTING
Preparing data and building forecasting models for 7-day sales prediction
"""

# Fix NumPy compatibility issue with statsmodels
%pip install --upgrade numpy==1.24.3 statsmodels==0.14.0 pmdarima==1.8.5

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Time series specific libraries
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX 


# Load data
df = pd.read_csv(r'C:\Users\hp\Downloads\archive (12)\Superstore_sales\train.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y', errors='coerce')

# Create daily sales aggregation
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
daily_sales.set_index('Order Date', inplace=True)

# Create monthly sales for decomposition
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum()

print("✅ Time series data prepared!")
print(f"Daily sales shape: {daily_sales.shape}")
print(f"Monthly sales shape: {monthly_sales.shape}")
```

## # Cell 2: Seasonal Decomposition

```python
import os

# Create the folder if it doesn't exist
os.makedirs('../visualizations', exist_ok=True)

# Now save the figure
plt.savefig('../visualizations/seasonal_decomposition.png', dpi=300, bbox_inches='tight')
```

```python
# Cell 2: Seasonal Decomposition
"""
## 🔄 SEASONAL DECOMPOSITION
Breaking down sales into trend, seasonal, and residual components
"""

# Perform seasonal decomposition
decomposition = seasonal_decompose(monthly_sales, model='additive', period=12)

# Plot decomposition
fig, axes = plt.subplots(4, 1, figsize=(14, 12))

# Original
axes[0].plot(monthly_sales.index, monthly_sales.values, color='#2E86AB')
axes[0].set_title('Original Sales Series', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Sales ($)')
axes[0].grid(True, alpha=0.3)

# Trend
axes[1].plot(monthly_sales.index, decomposition.trend, color='#A23B72')
axes[1].set_title('Trend Component', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Sales ($)')
axes[1].grid(True, alpha=0.3)

# Seasonal
axes[2].plot(monthly_sales.index, decomposition.seasonal, color='#5FAD56')
axes[2].set_title('Seasonal Component', fontsize=14, fontweight='bold')
axes[2].set_ylabel('Sales ($)')
axes[2].grid(True, alpha=0.3)

# Residual
axes[3].plot(monthly_sales.index, decomposition.resid, color='#F2C14E')
axes[3].set_title('Residual Component', fontsize=14, fontweight='bold')
axes[3].set_ylabel('Sales ($)')
axes[3].set_xlabel('Date')
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../visualizations/seasonal_decomposition.png', dpi=300, bbox_inches='tight')
plt.show()

print("=== DECOMPOSITION INSIGHTS ===")
print(f"Overall Trend: {'Declining' if decomposition.trend.iloc[-1] < decomposition.trend.iloc[0] else 'Increasing'}")
print(f"Seasonal Pattern: Clear yearly seasonality detected")
print(f"Residual Variance: {decomposition.resid.var():.2f}")
```

## # Cell 3: Stationarity Testing and ACF/PACF

```python
# Cell 3: Stationarity Testing and ACF/PACF
"""
## 📊 STATIONARITY ANALYSIS
Testing if the time series is stationary and analyzing autocorrelation
"""

# Augmented Dickey-Fuller test
result = adfuller(monthly_sales.dropna())
print('=== AUGMENTED DICKEY-FULLER TEST ===')
print(f'ADF Statistic: {result[0]:.4f}')
print(f'p-value: {result[1]:.4f}')
print(f'Critical Values:')
for key, value in result[4].items():
    print(f'\t{key}: {value:.4f}')

if result[1] <= 0.05:
    print("\n✅ Series is stationary (p-value <= 0.05)")
else:
    print("\n❌ Series is not stationary - will apply differencing")

# Plot ACF and PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

plot_acf(monthly_sales.dropna(), lags=23, ax=ax1)
ax1.set_title('Autocorrelation Function (ACF)', fontsize=14, fontweight='bold')

plot_pacf(monthly_sales.dropna(), lags=23, ax=ax2)
ax2.set_title('Partial Autocorrelation Function (PACF)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('../visualizations/acf_pacf_plots.png', dpi=300, bbox_inches='tight')
plt.show()

# First difference to make stationary
monthly_sales_diff = monthly_sales.diff().dropna()

# Test stationarity of differenced series
result_diff = adfuller(monthly_sales_diff)
print('\n=== AFTER DIFFERENCING ===')
print(f'ADF Statistic: {result_diff[0]:.4f}')
print(f'p-value: {result_diff[1]:.4f}')
```

## # Cell 4: ARIMA Model

```python
# Cell 4: ARIMA Model
"""
## 📈 ARIMA MODEL FOR SALES FORECASTING
Building ARIMA model to forecast next 7 days
"""

# Use auto_arima to find optimal parameters
print("Finding optimal ARIMA parameters...")
try:
    auto_model = auto_arima(monthly_sales, 
                           start_p=0, start_q=0,
                           max_p=5, max_q=5,
                           seasonal=False,
                           trace=True,
                           error_action='ignore',
                           suppress_warnings=True,
                           stepwise=True)
    
    print(f"\nOptimal ARIMA parameters: {auto_model.order}")
    best_order = auto_model.order
except:
    print("Auto ARIMA failed, using default (1,1,1)")
    best_order = (1, 1, 1)

# Fit ARIMA model
model_arima = ARIMA(monthly_sales, order=best_order)
model_fit = model_arima.fit()

print("\n=== ARIMA MODEL SUMMARY ===")
print(model_fit.summary())

# Forecast next 7 days (convert from monthly to daily approximation)
# For 7-day forecast, we'll use daily data
daily_sales_clean = daily_sales.dropna()

# Fit ARIMA on daily data for 7-day forecast
daily_model = ARIMA(daily_sales_clean, order=best_order)
daily_fit = daily_model.fit()

# Forecast next 7 days
forecast_steps = 7
forecast = daily_fit.forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=daily_sales_clean.index[-1] + timedelta(days=1), 
                               periods=forecast_steps, freq='D')

# Create forecast dataframe
forecast_df = pd.DataFrame({
    'Date': forecast_index,
    'Forecasted_Sales': forecast.values,
    'Lower_Bound': forecast.values * 0.85,  # 85% confidence interval
    'Upper_Bound': forecast.values * 1.15
})

print("\n=== 7-DAY SALES FORECAST (ARIMA) ===")
print(forecast_df)

# Plot forecast
fig, ax = plt.subplots(figsize=(14, 6))

# Plot historical data (last 60 days)
historical = daily_sales_clean.tail(60)
ax.plot(historical.index, historical.values, 
        label='Historical Sales', color='#2E86AB', linewidth=2)

# Plot forecast
ax.plot(forecast_index, forecast.values, 
        label='Forecast', color='#A23B72', linewidth=2, linestyle='--')

# Add confidence interval
ax.fill_between(forecast_index, 
                forecast_df['Lower_Bound'], forecast_df['Upper_Bound'],
                alpha=0.2, color='#A23B72', label='Confidence Interval')

ax.axvline(x=daily_sales_clean.index[-1], color='red', 
           linestyle='--', alpha=0.5, label='Present')

ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Sales ($)', fontsize=12)
ax.set_title('7-Day Sales Forecast - ARIMA Model', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../visualizations/arima_forecast_7days.png', dpi=300, bbox_inches='tight')
plt.show()
```

# # Cell 5: Prophet Model

```python
# Cell 5: Prophet Model
"""
## 🔮 FACEBOOK PROPHET MODEL
Using Prophet for advanced time series forecasting
"""

%pip install prophet

from prophet import Prophet

# Prepare data for Prophet
prophet_df = daily_sales_clean.reset_index()
prophet_df.columns = ['ds', 'y']  # Prophet requires specific column names

# Initialize and fit Prophet model
model_prophet = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
    changepoint_prior_scale=0.05,
    seasonality_prior_scale=10.0
)

# Add quarterly seasonality
model_prophet.add_seasonality(name='quarterly', period=90.5, fourier_order=5)

# Fit the model
model_prophet.fit(prophet_df)

# Create future dataframe for next 7 days
future = model_prophet.make_future_dataframe(periods=7)
forecast_prophet = model_prophet.predict(future)

# Extract last 7 days forecast
forecast_7days = forecast_prophet.tail(7)[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forecast_7days.columns = ['Date', 'Forecasted_Sales', 'Lower_Bound', 'Upper_Bound']

print("=== 7-DAY SALES FORECAST (PROPHET) ===")
print(forecast_7days)

# Plot Prophet forecast
fig1 = model_prophet.plot(forecast_prophet)
plt.title('Prophet Sales Forecast with 7-Day Prediction', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../visualizations/prophet_forecast_7days.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot components
fig2 = model_prophet.plot_components(forecast_prophet)
plt.tight_layout()
plt.savefig('../visualizations/prophet_components.png', dpi=300, bbox_inches='tight')
plt.show()

```
